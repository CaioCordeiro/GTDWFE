# Evaluating feature selection algorithms to apply in antimicrobial-resistant genes classification in Gram-negative bacterias.

Explore feature selection and evaluation algorithms to select the most important features of antimicrobial-resistant genes in Gram-negative bacterias.

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import cross_validate
from feature_selection.relieff_algorithm import apply_relieff
from comparators.scores import get_mean_scores
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/Ac_Sa_Ca_Kl_Ec/bla_all.csv")
```

```python
df.describe()
```

## Discretizing

```python
def discretize(df):
    for i in df.columns:
        if i != "Feature":
            df[i] = pd.qcut(df[i], q=5,  labels=False, precision=0, duplicates='drop')
    return df
```

```python
df = discretize(df)
```

```python
df.describe()
```

```python
df
```

## Dividing train and test data

```python
#df = pd.get_dummies(df, columns=["Feature"])
y = df["Output"]
X = df.drop(["Output","Feature"], axis=1) 
X_copy = X.copy(deep=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=4)
clf = svm.SVC()
#clf = clf.fit(X_train, y_train)
```

```python
X_copy = apply_relieff(X_copy.to_numpy(), y)
print(len(X_copy))
X_train2, X_test2, y_train2, y_test2 = train_test_split( X_copy, y, test_size=0.30, random_state=4)
clf2 = svm.SVC()
#clf2 = clf2.fit(X_train2, y_train2)
```

```python
disp = plot_confusion_matrix(clf, X_test, y_test,
                                 cmap=plt.cm.Blues)
disp.ax_.set_title("Confusion Matrix")
disp = plot_confusion_matrix(clf2, X_test2, y_test2,
                                 cmap=plt.cm.Blues)

plt.show()
```

```python
scores = cross_val_score(clf, X_test, y_test, cv=5, scoring='f1_macro')
print("%0.2f accuracy with a standard deviation of %0.2f on model 1" % (scores.mean(), scores.std()))
scores = cross_val_score(clf, X_test2, y_test2, cv=5, scoring='f1_macro')
print("%0.2f accuracy with a standard deviation of %0.2f on model 2" % (scores.mean(), scores.std()))
```

```python
y_pred = cross_val_predict(clf, X_test, y_test, cv=10)
y_pred2 = cross_val_predict(clf2, X_test2, y_test2, cv=10)
```

```python
y_results = cross_validate(clf, X, y, cv=10, return_train_score=True)
get_mean_scores(y_results)
```

```python
print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test2, y_pred2))
```
